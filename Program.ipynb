```python
import os
import zipfile
import shutil
import glob
```

```python
zip_file = os.listdir() #getting all the files in the directory as a list that i can iterate on
print(zip_file)
```

```python
os.mkdir("handwritten data")# making a folder to extract all the files in zipfile
```

```python
#program to unzip the zip files
for i in zip_file:
    unzip_zip = zipfile.ZipFile(i,"r") #Reads the zip file

    unzip_zip.extractall("./handwritten data") #Extracting the zipfile to a folder called handwritten data

    unzip_zip.close() #Closes the zipfile
```

```python
#scans all the folders and subfolders and creates a list of all paths of the jpeg files

files = glob.glob('./handwritten data/**/*.jpg',  
                   recursive = True) 
```

```python
type(files)
```

```python
len(files)
```

```python
#printing all the paths of the jpeg files
for file in files:
    print(file)
```

```python
#creating a new folder to store all the jpeg files
os.mkdir("All Data")
```

```python
#program to renames the jpeg files and store it to a specified folder
for count,file in enumerate(files): #iterating over the list of path names of the jpeg files
    dst = f"All Data/HW {str(count)}.jpg" #the destination folder and the name of the files 
    src = f'{file}' #the path of the source files  
    
    os.rename(src,dst) #function to rename the files
```
